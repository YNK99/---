```python
import requests
import pandas as pd
```

```python
url = 'http://apis.data.go.kr/1140100/CivilPolicyQnaService/PolicyQnaList?serviceKey='
key = '인증키'
A = '&regFrom=20001101&regTo=20210928&firstIndex='
B = '&recordCountPerPage=1000'      # 한 페이지당 민원 1000개를 불러옴
```

```python
print(url+key+A)
```

```python
data = []
cnt = 0
for i in range(50):
    cnt += 1
    re = requests.get(url + key + A + str(cnt) + B)
    for j in re.json()['resultList']:
        data.append(j)

    print(re)
    print(cnt)
```

```python
len(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("민원항목.csv", index = False)
```

```python
sea_bool = (df['ancName'] == '해양수산부')
```

```python
sea = df[sea_bool]
sea
```

```python
url = 'http://apis.data.go.kr/1140100/CivilPolicyQnaService/PolicyQnaItem?serviceKey='
duty = '&dutySctnNm=tqapttn'
```

```python
cnt = 0
s_qnacontent = []
for i in sea['faqNo']:
    try:
        print(i)
        re = requests.get(url + key + '&faqNo=' + str(i) + duty)
        s_qnacontent.append(re.json()['resultData'])
        cnt += 1
        print(cnt)
    except:
        continue
```

```python
s_qna = pd.DataFrame(s_qnacontent)
```

```python
s_qna.to_csv("해양수산부민원.csv", index = False)
```
